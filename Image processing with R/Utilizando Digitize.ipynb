# Extração de dados de um gráfico utilizando Digitize

    Outra alternativa encontrada para extrair os dados de um gráfico em uma figura, foi por meio do pacote para R 'Digitize'. Esse método é mais utilizado para recuperar dados plotados em um gráfico antigos feitos em R. Abaixo será apresentado o passo a passo desse pacote.
   

```r
install.packages('digitize', repos='http://cran.us.r-project.org')
```

```r
library(digitize)
cal = ReadAndCal('teste1.png')
data.points = DigitData(col = 'red')
```
